# Pruebas unitarias

Disciplina de probar los métodos

```python
suma = lambda x,y: x+y
```

Esta función puede ser validada con casos conocidos usando `assert`

```python
assert suma(2,2) == 4
```

Lanza un error pues no se cumple la igualdad

```python
assert suma(2,3) == 4
```

# Lanzar un error

Usando `raise`

```python
age = 10
if age >= 18:
    print("Permitir entrada")
else:
    raise Exception("No se permiten menores de edad")
```

- `assert()` hace una verificacion y manda un `AssertionError` cuando no se cumple el “statement”
- `raise Exception()`levanta un error creado por nosotros mismos.

# Manejo de excepciones

```python
print(0/0)
```

Manejo de error usando try

```python
try:
    print(0/0)
except ZeroDivisionError as error:
    print(error)
```

```python
try: 
    age = 10
    if age >= 18:
        print("Permitido ingresar.")
    else:
        raise Exception("No permitido ingreso de menores de edad")
except Exception as error: 
    print(error)

print('Fin de la ejecución.')
```

# Esquema General: Manejo de exepciones

```python
try:
    # Corre este código.
except:
    # Ejecutar este código cuando hay una excepción.
else:
    # ¿No hay excepciones? Corre este código.
finally:
    # Siempre corre este código.
```

```python
try:
  print(0/0)
  assert 1 != 1, 'Uno no es igual a uno'
  age = 10
  if age < 18:
    raise Exception('No se permite menores de edad')
except ZeroDivisionError as error:
  print(error)
except AssertionError as error:
  print(error)
except Exception as error:
  print(error)
```
